```python
def more_detils(item_id,cookies = None,headers=None):

    response = requests.get(f'https://www.yad2.co.il/api/item/{item_id}', cookies=cookies, headers=headers).json()
    details  = {}
    try:
        
        details['price'] = response['price']
        details['street'] = response['street']
        details['neighborhood'] = response['neighborhood']
        details['parking'] = response['parking']
        details['balconies'] = response['balconies']
        details['buildingMR'] = response['square_meters']
        details['TotalFloors'] = response['TotalFloor_text']
        details['asset_classification'] = response['analytics_items']['asset_classification']
        details['rooms'] = response['analytics_items']['rooms']
        details['floor'] = response['analytics_items']['floor']
        details['shelter'] = response['analytics_items']['shelter_room']
        details['on_pillars'] = response['analytics_items']['on_pillars']
        details['elevator'] = response['analytics_items']['elevator']
        details['storeroom'] = response['analytics_items']['storeroom']

        details['lat']  = response['analytics_items']['lat']
        details['long']  = response['analytics_items']['long']
        
        details['home_number'] = response['address_home_num']
        details['home_number_2'] = response['address_home_number']
            
        
    except:
        pass 

    
    return details


def find_number_of_pages(cookies=None,headers=None):
    params = {
            'topArea': '2',
            'area': '1',
            'city': '5000',
            'propertyGroup': 'apartments',
            'price': '600000--1',
            'page': 1,
        }
    response = requests.get('https://www.yad2.co.il/realestate/forsale', params=params , cookies=cookies, headers=headers)
    soup = BeautifulSoup(response.content, 'html5lib')
    table = soup.find('div', attrs = {'class':'column_large'}) 
    if soup.title.text == 'ShieldSquare Captcha':
        return print(soup.title.text)
    elif str(response) !='<Response [200]>':
        return print(f'-Error:---{response}----')
    
    return int(table.find('button', attrs = {'class':'page-num'}).get_text(strip = True))
    
```
